```python
length_capture = 20
overlap_frame = 4
start_at = 7
end_at = 100

DEBUG = True
captured = []
frame=[]
captured_frame_count = 0
the_overlap=[]

for i in range(1,100):
    if(i>start_at-1 and i<end_at+1):
        if i==end_at-1:
            captured.append(frame)
        frame.append(i)
        if len(frame)>=(length_capture-overlap_frame) and len(frame)<=length_capture:
            #frame.append(i)
            the_overlap.append(i)
            if DEBUG: print('f:',frame)
            if DEBUG: print('o:',frame)
        if len(frame)>length_capture:
            the_overlap.append(i)
            frame.pop()
            captured.append(frame)
            frame=[]
            frame.extend(the_overlap)
            the_overlap=[]
            if DEBUG: print('cc:',captured)
```
